```python

def generate_structured_points(num_points=1000):
    grid_size = int(np.sqrt(num_points))
    x, y = np.meshgrid(np.linspace(0, 1, grid_size), np.linspace(0, 1, grid_size))
    points = np.vstack([x.ravel(), y.ravel()]).T
    colors = points[:, 0]  # Farbverlauf basierend auf der x-Koordinate
    return points, colors

def apply_combined_transformation(points, tx, ty, sx, sy, angle):
    translation_matrix = np.array([[1, 0, tx], [0, 1, ty], [0, 0, 1]])
    scaling_matrix = np.array([[sx, 0, 0], [0, sy, 0], [0, 0, 1]])
    rotation_matrix = np.array([[np.cos(angle), -np.sin(angle), 0], [np.sin(angle), np.cos(angle), 0], [0, 0, 1]])
    transformation_matrix = translation_matrix @ scaling_matrix @ rotation_matrix
    ones = np.ones((points.shape[0], 1))
    points_with_ones = np.hstack([points, ones])
    transformed_points = points_with_ones @ transformation_matrix.T
    return transformed_points[:, :2], transformation_matrix

def interactive_plot(tx=0.0, ty=0.0, sx=1.0, sy=1.0, angle=0.0):
    transformed_points, transformation_matrix = apply_combined_transformation(structured_points, tx, ty, sx, sy, angle)
    plt.scatter(transformed_points[:, 0], transformed_points[:, 1], c=structured_colors, cmap='hsv', alpha=0.6)
    plt.xlim(0, 2)
    plt.ylim(0, 2)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()
    print("Transformationsmatrix:\n", transformation_matrix)
```

```python
structured_points, structured_colors = generate_structured_points(1000)

interact(interactive_plot,
         tx=FloatSlider(min=-1.0, max=1.0, step=0.1),
         ty=FloatSlider(min=-1.0, max=1.0, step=0.1),
         sx=FloatSlider(min=0.5, max=2.0, step=0.1),
         sy=FloatSlider(min=0.5, max=2.0, step=0.1),
         angle=FloatSlider(min=-np.pi, max=np.pi, step=0.1));
```
