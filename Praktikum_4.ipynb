# Praktikum 4: Attention und Transformer

```python
# @title 1 Softmax auf Aufmerksamkeitsmatrix Implementieren und Analysieren

import torch
import torch.nn as nn
import math
import seaborn as sns
import matplotlib.pylab as plt
plt.rcParams['figure.figsize'] = [20, 7]

n = 30 # @param {type:"slider", min:2, max:30, step:1}
d = 1024 # @param {type:"slider", min:2, max:1024, step:1}

Q = torch.randn(n, d)
K = torch.randn(n, d)

attention_matrix_1 = torch.softmax(Q @ K.transpose(0, 1) / math.sqrt(d), dim=-1)
attention_matrix_2 = torch.softmax(Q @ K.transpose(0, 1), dim=-1)

fig, (ax1, ax2) = plt.subplots(1, 2)
sns.heatmap(attention_matrix_1, linewidth=0.5, vmin=0, vmax=1, ax=ax1)
sns.heatmap(attention_matrix_2, linewidth=0.5, vmin=0, vmax=1, ax=ax2)
plt.show()
```
